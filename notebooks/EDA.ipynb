```python
# Importamos Librerias
import pandas as pd
import streamlit as st
import plotly.express as px

import plotly.io as pio
pio.templates.default = "plotly_white"
```

```python
# Importamos los datos
df = pd.read_csv("..\housing_data_CDMX_v2.csv")

# Mostramos informacion general del DataFrame
df.info()
display(df.head())
```

> Como podemos observar no tenemos valores nulos, las columnas tienen el formato correcto y el estilo de los encabezados es adecuado. Quitaremos solo las columnas de coordenasas al no ser de utilidad para nuestro analisis.

```python
# Caculamos el tipo de cambio promedio para MXN y USD
tipo_cambio = (df['price_aprox_local_currency'] / df['price_aprox_usd']).mean()
print(f"Tipo de cambio promedio (MXN/USD): {tipo_cambio}")
```

```python
# Convertimos a pesos mexicanos los precios en USD
df['price'] = df['price'].where(df['currency'] == 'MXN', df['price'] * tipo_cambio)
df['price_per_m2'] = df['price_per_m2'].where(df['currency'] == 'MXN', df['price_per_m2'] * tipo_cambio)
```

```python
# Borramos las columnas lat, lon y lat-lon
df = df.drop(columns=['lat', 'lon', 'lat-lon'])

# Borramos las columnas con datos en USD
df = df.drop(columns=['price_aprox_usd', 'price_usd_per_m2'])
```

```python
# Mostramos datos estadisticos del DataFrame
display(df.describe())
```

```python
# Mostramos los 20 registros ocn mayor superficie total
display(df.nlargest(20, 'surface_total_in_m2'))
```

> Usaremos el precio por metro cuadrado como indicador para filtrar la tabla

```python
print('Tamaño del DF antes de limpiar outliers: ', df.shape)

# Eliminamos los outliers de las columnas numericas
numeric_cols = ['price_per_m2', 'surface_total_in_m2']
for col in numeric_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
# Mostramos informacion del DataFrame despues de limpiar outliers
print('Tamaño del DF despues de limpiar outliers: ', df.shape)
```

# Graficos

```python
min_price = df['price_aprox_local_currency'].min()
max_price = df['price_aprox_local_currency'].max()

#max_price = df['price_aprox_local_currency'].quantile(0.85)  # Usamos el percentil 75 para evitar outliers extremos

print(f"Rango de precios considerado para el análisis: {min_price} - {max_price}")

df_filtered = df[(df['price_aprox_local_currency'] >= min_price) & (df['price_aprox_local_currency'] <= max_price)]

# Creamos un gráfico para visualizar la distribución de precios x tipo de vivienda
fig = px.histogram(df_filtered, x='price_aprox_local_currency', nbins=50, title='Distribución de Precios de Vivienda en CDMX',
                   labels={'price_aprox_local_currency': 'Precio (MXN)'}, range_x=[min_price, max_price], color='property_type')

fig.update_layout(yaxis_title='Cantidad de Propiedades') # Cambiamos el nombre del eje y
fig.update_layout(legend_title_text='Tipo de Vivienda') # Cambiamos el nombre de las leyendas

fig.show()
```

```python
# Cremaos una gráfico para mostrar el precio promedio por tipo de vivienda
avg_price_by_type = df_filtered.groupby('property_type')['price_aprox_local_currency'].mean().reset_index()
fig2 = px.bar(avg_price_by_type, x='property_type', y='price_aprox_local_currency',
              title='Precio Promedio por Tipo de Vivienda en CDMX',
              labels={'property_type': 'Tipo de Vivienda', 'price_aprox_local_currency': 'Precio Promedio (MXN)'})
fig2.update_layout(yaxis_title='Precio Promedio (MXN)') # Cambiamos el nombre del eje y
fig2.show()
```

```python
# Grafico precio promedio por delegación
avg_price_by_place = df_filtered.groupby('places')['price_aprox_local_currency'].mean().reset_index()
fig3 = px.bar(avg_price_by_place, x='places', y='price_aprox_local_currency',
              title='Precio Promedio por Delegación en CDMX',
              labels={'places': 'Delegación', 'price_aprox_local_currency': 'Precio Promedio (MXN)'})
fig3.update_layout(yaxis_title='Precio Promedio (MXN)') # Cambiamos el nombre del eje y
fig3.show()
```

```python
# Histograma de precios por delegación
fig3 = px.histogram(df_filtered, x='price_aprox_local_currency', nbins=50, title='Distribución de Precios por Delegación en CDMX',
                   labels={'price_aprox_local_currency': 'Precio (MXN)'}, range_x=[min_price, max_price], color='places')
fig3.update_layout(yaxis_title='Cantidad de Propiedades') # Cambiamos el nombre del eje y
fig3.update_layout(legend_title_text='Delegación') # Cambiamos el nombre de las leyendas
fig3.show()
```

```python
#Diagrama de dispersion entre precio y superficie total
fig4 = px.scatter(df_filtered, x='surface_total_in_m2', y='price_aprox_local_currency',
                  title='Relación entre Superficie Total y Precio en CDMX',
                  labels={'surface_total_in_m2': 'Superficie Total (m²)', 'price_aprox_local_currency': 'Precio (MXN)'})
fig4.update_layout(yaxis_title='Precio (MXN)') # Cambiamos el nombre del eje y
fig4.update_layout(xaxis_title='Superficie Total (m²)') # Cambiamos el nombre del eje x
fig4.show()
```

```python
# Diagrama de dispersion entre precio por metro cuadrado y superficie total
fig5 = px.scatter(df_filtered, x='surface_total_in_m2', y='price_per_m2',
                  title='Relación entre Superficie Total y Precio por Metro Cuadrado en CDMX',
                  labels={'surface_total_in_m2': 'Superficie Total (m²)', 'price_per_m2': 'Precio por Metro Cuadrado (MXN)'})
fig5.update_layout(yaxis_title='Precio por Metro Cuadrado (MXN)') # Cambiamos el nombre del eje y
fig5.update_layout(xaxis_title='Superficie Total (m²)') # Cambiamos el nombre del eje x
fig5.show()
```
